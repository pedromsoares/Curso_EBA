#Corrigindo campo milliseconds df results

```python
results['milliseconds']=results['milliseconds'].replace(0,np.mean(results['milliseconds']))
```

```python
df_races_status=(results
                 .merge(races,on='raceId')
                 .merge(status, on='statusId')
                 .merge(circuits,on='circuitId'))
```

```python
df_races_status.head()
```

```python
status_circuits=(df_races_status
                 .groupby(['status','name_y'])
                 ['driverId']
                 .count()
                 .to_frame()
                 .reset_index())
```

```python
status_circuits.head()
```

```python
status_circuits['status'].unique()
```
