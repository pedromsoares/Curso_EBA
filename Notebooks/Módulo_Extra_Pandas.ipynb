### Groupby

```python
df.groupby('artist').count()
```

```python
df.groupby('artist')['title'].count()
```

```python
df.groupby(['artist', 'year'])['title'].count() #Agrupamento com a lista das colunas
```

```python
df.groupby(['artist', 'year']).agg({'year':'count', 'dur':'mean'}) #Faço um agrupamento pelo artista e o ano e realizo a contagem do ano e a média da duração
```

```python
df_group_year = df.groupby('year') #novo dataframe com agrupamento por ano
```

```python
df_group_year.ngroups #Retorna o número de grupos criados
```

```python
df_group_year.groups #Retorna um dicionário com os grupos (anos) e os índíces dos dados
```

```python
df_group_year.size() #Retorna o tamanho dos grupos
```

```python
df_group_year.get_group(2015) #Retorna o grupo de 2015
```

## Merging Dataframes

O .merge() funciona como o left join do SQL.

inner-join-left-join-outer-join.png

```python
df.merge(df2, left_on='artist', right_on='Artists', how='left').fillna(0)
```

```python
df.merge(df2, left_on='artist', right_on='Artists', how='left', indicator=True) #Une os dataframes como antes, mas o parâmetro indicator adiciona uma coluna dizendo em qual dataframe a variável estava
```

### Concatenando operações

```python
df_grouped = (df
              .groupby('artist')
              ['title']
              .count()
              .to_frame()
              .reset_index()
              )
```
